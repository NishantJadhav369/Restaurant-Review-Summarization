```python
def extract_data(page_url):
    response = requests.get(page_url)
    html_file = response.text
    soup = BeautifulSoup(html_file,'html.parser')
    
    rname = soup.findAll('h1')[1].contents
    d0 = pd.DataFrame(rname,columns=['Name'])
    rtitle=[]
    for span in soup.findAll('span','noQuotes'):
        rtitle.append(span.contents[0])
    rdesc = []
    for p in soup.findAll('p','partial_entry'):
        rdesc.append(p.contents[0])
    d1 = pd.DataFrame(rtitle,columns=['Title'])
    d2 = pd.DataFrame(rdesc,columns=['Desc'])
    d3 = d1.join(d2)
    d4 = d3.join(d0)
    #d4.head()
    f1 = d4.fillna(d4['Name'].iloc[0])
    d4.head()
    return f1
    #print(f1)
```

```python
def hotel_link(start):
    root_link = 'https://www.tripadvisor.in'+start
    response = requests.get(root_link)
    html_file = response.text
    soup = BeautifulSoup(html_file,'html.parser')
    pages = soup.findAll('div','pageNumbers')
    if(len(pages)==0):
        return
    link = pages[0].findAll('a',href=True)
    lnew = []
    links =[]
    for tag in link:
        #print(tag['href'])
        links.append(tag['href'])
    ldummy = pd.DataFrame(columns=['Title','Desc','Name'])
    for i,f in enumerate(links,0):
        lf = extract_data('https://www.tripadvisor.in/'+links[i][1:])
        lnew= pd.concat([ldummy,lf])
        lnew = lnew.reset_index().drop('index',axis=1)
        ldummy = lnew
    return lnew
```

```python
def new_page(page):
    start_url = 'https://www.tripadvisor.in'+page
    response = requests.get(start_url)
    rsoup = BeautifulSoup(response.text, "html.parser")
    rlinks = rsoup.findAll('a',href=True)
    starts =[]
    for atag in rsoup.findAll('a','restaurants-list-ListCell__restaurantName--2aSdo'):
        starts.append(atag['href'])
    dfs =[]
    for head in starts:
        data = hotel_link(head)
        dfs.append(data)
    final = pd.concat(dfs).reset_index().drop('index',axis=1)
    return final
    
```

```python
def hotel_link_new(start):
    #root_link = 'https://www.tripadvisor.in'+start
    response = requests.get(start)
    html_file = response.text
    soup = BeautifulSoup(html_file,'html.parser')
    pages = soup.findAll('div','pageNumbers')
    if(len(pages)==0):
        return
    link = pages[0].findAll('a',href=True)
    lnew = []
    links =[]
    for tag in link:
        #print(tag['href'])
        links.append(tag['href'])
    ldummy = pd.DataFrame(columns=['Title','Desc','Name'])
    for i,f in enumerate(links,0):
        lf = extract_data('https://www.tripadvisor.in/'+links[i][1:])
        lnew= pd.concat([ldummy,lf])
        lnew = lnew.reset_index().drop('index',axis=1)
        ldummy = lnew
    return lnew
```
